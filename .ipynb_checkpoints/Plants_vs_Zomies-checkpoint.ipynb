```c++
//include all needet Liberys
#include <algoviz/SVG.hpp>
#include <string>
#include <math.h>
#include <sstream>

#include <chrono>
#include <iomanip>
#include <numeric>


//define the size of Objekts 
#define TILE_SIZE 40
#define TILE_COUNT_X 9
#define TILE_COUNT_Y 5

#define OFFSET_BOTTOM 40

#define KILLD_ZOMBIES_TO_WIN 5 

//Eingene Libery für Spielfunktionen
#include "Lib/Manegment.hpp"



//namespace Definieren
using namespace std;
using namespace std::chrono;

//Create the main Playfield Playground
AlgoViz::clear();
AlgoViz::sendText("Game Startet"); 
SVG Playground = SVG (TILE_SIZE*TILE_COUNT_X,(TILE_SIZE*TILE_COUNT_Y + OFFSET_BOTTOM));


//input / Output
MouseState maus;
string taste;

//Timing

//Game clock für 30FPS
#define GLOBAL_GAME_CLOCK 30

//Variable für die Zeit Deffineiern
double time_temp = 0.0;
double time_temp_offset = 0.0;

long last_30th_sec = 0;
long time1_int = 0;
typedef std::chrono::high_resolution_clock::time_point TimeVar;

//Abkürzung für Komplexe Befehle
#define timeNow() std::chrono::high_resolution_clock::now()

TimeVar time1  = timeNow();
TimeVar tstart = timeNow();
long counter_1 = 0;

//Startkontostand
int algos = 10;

//Zombie Spwan Mode
int counter_2 = 0;
int zombie_mode = 0;


//Pflanzenauswahl
int Plant_cost[4] = {10,4,17,23};
int current_plant = 0;
```

```c++
Manegment manegment = Manegment(&Playground,KILLD_ZOMBIES_TO_WIN);
Plant plant_to_show_of[4];
```

```c++
void draw_playground(){
    
    //spielfeld zeichen
    for(int x = 0; x < TILE_COUNT_X; x++){
        for(int y = 0; y < TILE_COUNT_Y; y++){
            if(x == 0){
                Playground.setFill(255,127,36);
            }else if((x%2 == 1 && y%2 == 1)||(x%2 == 0 && y%2 == 0)){
                Playground.setFill(0,200,0);
            }else {
                Playground.setFill(0,255,0);
            }
            Playground.drawRect(x*TILE_SIZE,(y*TILE_SIZE),TILE_SIZE,TILE_SIZE);
        }
    }
    
    //auswahl zeichnen
    for(int x = 0;x < 4; x++){
        int position_x = TILE_SIZE * (x+2);
        position_x = position_x +(TILE_SIZE /2);
        
        int position_y = TILE_SIZE*TILE_COUNT_Y;
        position_y = position_y + (TILE_SIZE / 2);
        
        plant_to_show_of[x] = Plant(position_x,position_y,x,&Playground);
    }
}
```

```c++
int get_kliked_TILE_x(){
    return maus.x()/TILE_SIZE;
}
```

```c++
int get_kliked_TILE_y(){
    return maus.y()/TILE_SIZE;
}
```

```c++
void change_algos (int a, int zombie_nr, int tote_Zombies){
    zombie_nr = zombie_nr+1;
    ostringstream outStream;
    outStream << "Algos: " << a << "  Zombie Type:" << zombie_nr << "   Tote Zombies:" <<tote_Zombies;
    Playground.clearStatusMsg();
    Playground.setStatusMsg (outStream.str());
}
```

```c++
void spawn_random_Zombies(int mode){
    switch(mode){
        case 0:
            if (rand() % 400 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 1:
            if (rand() % 350 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 2:
            if (rand() % 320 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 3:
            if (rand() % 300 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 4:
            if (rand() % 280 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 5:
            if (rand() % 270 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 6:
            if (rand() % 260 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 7:
            if (rand() % 260 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 8:
            if (rand() % 250 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 9:
            if (rand() % 250 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
        case 10:
            if (rand() % 240 == 42 ){
            manegment.Add_Zombie(((rand()%5)*TILE_SIZE)+(TILE_SIZE/2),0);
            }
        break;
            
        }
    }
```

```c++
int main(){
    draw_playground();
    change_algos(algos,current_plant,manegment.get_zombies_killed());
    do{
        //Timing
        time1 = timeNow();
                
        std::chrono::duration<double> diff = time1-tstart;
        
        time_temp = diff.count();
        time1_int = time_temp * GLOBAL_GAME_CLOCK;        
        
        if(last_30th_sec != time1_int){
            //Transmit how many 1/30sec are passd
            manegment.Move_movable_Objekts(time1_int-last_30th_sec);
            last_30th_sec = time1_int;
            counter_1++;
            if(counter_1>60){
                //etwa alle 3 sekunden
                counter_1 = 0;
                algos++;
                change_algos(algos,current_plant,manegment.get_zombies_killed());
            }
            //Spwan random Zombie
            counter_2++;
            if(counter_2 > 9000){
                //etwa alle 15 sekunden
                if(zombie_mode < 10){
                    zombie_mode = zombie_mode+1;
                }
            }
            spawn_random_Zombies(0);
        }
        
        
        
        //You Won the Game! now Interrupt the game
        if(manegment.Game_finished() == true){
            int do_play_endless = true;
            AlgoViz::sendText("You won the Game, type a for freeplay");                
            while (do_play_endless){
                AlgoViz::sleep(10);
                taste = Playground.lastKey();
                if(taste == "a"){
                    taste = "";
                    do_play_endless = false;
                }
            }
        }
        
        //You lost the game :-(
        
        if(manegment.Game_lost()>0){
            //You lost the game... do better next time
            AlgoViz::sendText("You Lost the Game");
            while(true){
                AlgoViz::sleep(100);
            }
        }
        
        //mauseingaben
        taste = Playground.lastKey();
        maus = Playground.lastClick();
        if (maus.isLegal() && maus.y() < TILE_SIZE*TILE_COUNT_Y && algos >= Plant_cost[current_plant]) {
            //testen ob an der Position noch keine Pflanze ist
            if(!manegment.get_plant_at_this_position((get_kliked_TILE_x()*TILE_SIZE)+(TILE_SIZE/2),(get_kliked_TILE_y()*TILE_SIZE)+(TILE_SIZE/2))){
                //Pflanze erstellen
                manegment.Add_Plant((get_kliked_TILE_x()*TILE_SIZE)+(TILE_SIZE/2),(get_kliked_TILE_y()*TILE_SIZE)+(TILE_SIZE/2),current_plant);
                
                //algos abziehen
                algos = algos - Plant_cost[current_plant];
                
                //akualisiere die anzeige
                change_algos(algos,current_plant,manegment.get_zombies_killed());
            }
        }
        else{
            //möglicherweise wurde eine Pflanze ausgewählt.
            int position_y = TILE_COUNT_Y;
            //position_y = position_y + 1;
            if(get_kliked_TILE_y() == position_y){
                current_plant = get_kliked_TILE_x();
                current_plant = current_plant-2;
                change_algos(algos,current_plant,manegment.get_zombies_killed());
            }
            //einfache auswahl unter dem Feld
        }

    }while(taste != "x");

}

```

```c++
main();
```
